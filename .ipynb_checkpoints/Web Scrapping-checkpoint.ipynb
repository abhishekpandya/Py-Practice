# Extracting Data using Web Scrapping

#### imports

```python
import requests
import bs4
```

```python
from requests import get
url = "http://example.webscraping.com/"    
response = get(url)
```

```python
html_str = response.text
```

```python
from IPython.core.display import display,HTML
display(HTML(response.text))
```

```python
from bs4 import BeautifulSoup
ps =  BeautifulSoup(html_str)
```

```python
# print ps
print(ps)
```

```python
# find content of the body tag
body = ps.find(name="body")
```

```python
print(body)
```

```python
# print all div elements
print(body.findAll(name="div"))
```

```python
# loop thru each element
for item in body.findAll(name="div"):
    print(item.text)
```

```python
# find tag by attributes for ex: id
print(body.findAll(name="div",  attrs={"id":"pagination"}))
```

```python
table = ps.find(name="table")
```

```python
for item in table:
    print(item.text)
```
